This notebook contains three maps created using **Cartopy**. Each map uses an appropriate projection and includes gridlines and labels for clarity.

### Import necessary Libraries

```python
import cartopy.crs as ccrs  # Projections
import cartopy.feature as cfeature  # Map features
import matplotlib.pyplot as plt  # Plotting
```

<h3 style="color: lightblue;">1. Antarctica and the Southern Ocean</h3>
<p>This map uses the <b>South Polar Stereographic projection</b> because it accurately represents high-latitude regions.</p>

```python
# Create a figure and axis with a South Polar Stereographic projection
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': ccrs.SouthPolarStereo()})

# Set the extent to show Antarctica and the Southern Ocean from 60°S
ax.set_extent([-180, 180, -90, -60], crs=ccrs.PlateCarree())

# Add coastlines and a blue ocean background
ax.coastlines(resolution='110m')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.add_feature(cfeature.LAND, color='navajowhite')

# Add gridlines
gl = ax.gridlines(draw_labels=True)
gl.right_labels = False  # Hide right labels
gl.top_labels = False  # Hide top labels

# Add a title
ax.set_title("Antarctica and the Southern Ocean", fontsize=14, fontweight="bold")

# Show the map
plt.show()
```

**Figure 1: Antarctica and the Southern Ocean**  
This map uses the **South Polar Stereographic projection** because it accurately represents high-latitude regions, minimizing distortion near the poles. It is ideal for visualizing polar regions, such as Antarctica and the Southern Ocean.


<h3 style="color: lightblue;">2. South Atlantic with Key Locations</h3>
<p>This map covers <b>20°S to 50°S</b> and includes important cities: Walvis Bay, Cape Town, Rio de Janeiro, and Montevideo.</p>

```python
# Import additional Cartopy features
import cartopy.feature as cfeature  

# Create a figure with PlateCarree projection
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': ccrs.PlateCarree()})

# Set the extent for the South Atlantic
ax.set_extent([-60, 20, -50, -20], crs=ccrs.PlateCarree())

# Add coastlines and gridlines
ax.coastlines(resolution='110m')
ax.gridlines(draw_labels=True)

# Add a light blue ocean and light green land
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.add_feature(cfeature.LAND, color='lightgreen')

# Define locations (longitude, latitude)
locations = {
    "Walvis Bay": (14.505, -22.957),
    "Cape Town": (18.424, -33.924),
    "Rio de Janeiro": (-43.172, -22.906),
    "Montevideo": (-56.164, -34.901)
}

# Add locations to the map
for city, (lon, lat) in locations.items():
    ax.plot(lon, lat, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    ax.text(lon + 1, lat, city, transform=ccrs.PlateCarree(), fontsize=10)

# Add a title
ax.set_title("South Atlantic with Key Locations", fontsize=14, fontweight="bold")

# Show the map
plt.show()
```

**Figure 2: South Atlantic with Key Locations**  
The **Plate Carrée projection** is used because it provides a simple cylindrical view, making it easy to plot key locations without distortion. This is useful for understanding the relative positions of cities along the South Atlantic Ocean.


<h3 style="color: lightblue;">False Bay - Coastline Resolutions</h3>
<p>This section compares <b>coarse, intermediate, and full</b> coastline resolutions.</p>

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Define the coastline resolutions
resolutions = ['110m', '50m', '10m']
titles = ['Coarse', 'Intermediate', 'Full']

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5), subplot_kw={'projection': ccrs.PlateCarree()}, constrained_layout=True)

for ax, res, title in zip(axes, resolutions, titles):
    # Set the extent for False Bay (around Cape Town)
    ax.set_extent([17, 19, -35.5, -34], crs=ccrs.PlateCarree())
    
    # Add coastlines with different resolutions
    ax.coastlines(resolution=res)
    
    # Add gridlines
    gl = ax.gridlines(draw_labels=True)
    gl.right_labels = False  # Hide right labels to reduce clutter
    
    # Set title for each subplot
    ax.set_title(f"False Bay - {title} Resolution")

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the figure
plt.show()
```

**Figure 3: False Bay - Coastline Resolutions**  
This comparison of **coarse (110m), intermediate (50m), and full (10m) coastline resolutions** demonstrates the differences in coastline detail. The **higher the resolution, the more detailed the coastal features**, which is important for navigation and coastal studies.
